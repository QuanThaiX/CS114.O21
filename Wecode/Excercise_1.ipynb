```python
def generate_subsets(n):
    subsets = [[]]
    for i in range(n):
        new_subsets = []
        for subset in subsets:
            new_subsets.append(subset + [i])
        subsets.extend(new_subsets)
    subsets.remove(subsets[0])
    return subsets

def check_replacement_columns(grid, columns, target_black_count, height, width):
    temp_grid = [row[:] for row in grid]
    for col in columns:
        for row in range(height):
            temp_grid[row][col] = "."
    black_count = sum([row.count("#") for row in temp_grid])
    return black_count == target_black_count

def check_replacement_rows(grid, rows, target_black_count, height, width):
    temp_grid = [row[:] for row in grid]
    for row in rows:
        for col in range(width):
            temp_grid[row][col] = "."
    black_count = sum([row.count("#") for row in temp_grid])
    return black_count == target_black_count

def check_replacement(grid, rows, columns, target_black_count, height, width):
    temp_grid = [row[:] for row in grid]
    for row in rows:
        for col in range(width):
            temp_grid[row][col] = "."
    for col in columns:
        for row in range(height):
            temp_grid[row][col] = "."
    black_count = sum([row.count("#") for row in temp_grid])
    return black_count == target_black_count

H, W, K = map(int, input().split())
row_subsets = generate_subsets(H)
column_subsets = generate_subsets(W)
grid = [list(input()) for _ in range(H)]
valid_combinations_count = 0
if sum([row.count("#") for row in grid]) == K:
    valid_combinations_count += 1
for row_subset in row_subsets:
    if check_replacement_rows(grid, row_subset, K, H, W):
        valid_combinations_count += 1
for column_subset in column_subsets:
    if check_replacement_columns(grid, column_subset, K, H, W):
        valid_combinations_count += 1
for row_subset in row_subsets:
    for column_subset in column_subsets:
        if check_replacement(grid, row_subset, column_subset, K, H, W):
            valid_combinations_count += 1
print(valid_combinations_count)
```

```python
def is_winner(board):
    for row in range(3):
        if board[row][0] + board[row][1] + board[row][2] == -3:
            return True
    for col in range(3):
        if board[0][col] + board[1][col] + board[2][col] == -3:
            return True
    if board[0][0] + board[1][1] + board[2][2] == -3 or board[0][2] + board[1][1] + board[2][0] == -3:
        return True
    return False

def play_loto_3x3(board, numbers):
    for number in numbers:
        for row in range(3):
            for col in range(3):
                if board[row][col] == number:
                    board[row][col] = -1
        if is_winner(board):
            return True
    return False

board = []
for _ in range(3):
    row = list(map(int, input().split()))
    board.append(row)
N = int(input())
drawn_numbers = [int(input()) for _ in range(N)]
if play_loto_3x3(board, drawn_numbers):
    print("Yes")
else:
    print("No")
```

```python
import math

def is_special_number(N, prime_factors):
    if N % 2 == 0:
        N //= 2
        if N % 2 == 0:
            return False
        prime_factors.add(2)
    i = 3
    while i <= math.isqrt(N):
        if N % i == 0:
            if i in prime_factors:
                return False
            else:
                prime_factors.add(i)
                N //= i
        else:
            i += 2
            if i % 3 == 0:
                i += 2
    if N > 2:
        if N in prime_factors:
            return False
        prime_factors.add(N)
    return True

def count_special_key_pairs(L, R):
    special_key_sets = []
    for number in range(L, R + 1):
        prime_factors = set()
        if is_special_number(number, prime_factors):
            special_key_sets.append(prime_factors)
    total_pairs = 0
    if L > 999:
        for i in range(len(special_key_sets) - 1):
            if len(special_key_sets[i]) == 1:
                total_pairs += (len(special_key_sets) - i - 1)
                continue
            for j in range(i + 1, len(special_key_sets)):
                if not special_key_sets[i].intersection(special_key_sets[j]):
                    total_pairs += 1
    else:
        for i in range(len(special_key_sets) - 1):
            for j in range(i + 1, len(special_key_sets)):
                if not special_key_sets[i].intersection(special_key_sets[j]):
                    total_pairs += 1

    return total_pairs

L, R = map(int, input().split())
result = count_special_key_pairs(L, R)
print(result)
```

```python
def pr_(a, b, c, d):
    print(f"({a[0]},{a[1]})", end=" ")
    print(f"({b[0]},{b[1]})", end=" ")
    print(f"({c[0]},{c[1]})", end=" ")
    print(f"({d[0]},{d[1]})", end="\n")

def pr(a, b, c, d):
    print(a, end=" ")
    print(b, end=" ")
    print(c, end=" ")
    print(d, end="\n")

A = tuple(map(int, input().split()))
B = tuple(map(int, input().split()))
dX = B[0] - A[0]
dY = B[1] - A[1]
C1 = (B[0] - dY, B[1] + dX)
D1 = (A[0] - dY, A[1] + dX)
C2 = (B[0] + dY, B[1] - dX)
D2 = (A[0] + dY, A[1] - dX)
if A[0] == B[0] and A[1] < B[1]:
    pr(A, B, C2, D2)
    pr(A, D1, C1, B)
elif A[0] <= B[0]:
    pr(A, D1, C1, B)
    pr(A, B, C2, D2)
else:
    pr(A, B, C2, D2)
    pr(A, D1, C1, B)
```

```python
import re

online_prices = list(map(int, input().split()))
price_descriptions = []
for _ in range(len(online_prices)):
    price_descriptions.append(input())
available_budget = int(input())
if sum(online_prices) <= available_budget:
    print("true")
else:
    in_store_prices = [0.0] * len(online_prices)
    for i in range(len(online_prices)):
        flag = 0
        discount_match = re.search(r'(\d+\.\d+)% lower than in-store', price_descriptions[i])
        discount_percentage = 0.0
        if discount_match:
            discount_percentage = float(discount_match.group(1))
            flag = -1
        else:
            increase_match = re.search(r'(\d+\.\d+)% higher than in-store', price_descriptions[i])
            if increase_match:
                discount_percentage = float(increase_match.group(1))
                flag = 1
        if flag == -1:
            in_store_prices[i] = (online_prices[i] * 100) / (100 - discount_percentage)
        elif flag == 1:
            in_store_prices[i] = (online_prices[i] * (100 + discount_percentage)) / 100
        else:
            in_store_prices[i] = online_prices[i]
    if sum(in_store_prices) <= available_budget:
        print("true")
    else:
        print("false")
```

```python
# 4x-6x faster ver

def prime_factorization(n):
    prime_factors = dict()
    i = 2
    while n > 1:
        if n % i == 0:
            if i not in prime_factors:
                prime_factors[i] = 1
            else:
                prime_factors[i] += 1
            n //= i
        else:
            i += 1
    if len(prime_factors) == 0:
        prime_factors[1] = 1
    return prime_factors

def extract_alpha_list(prime_factors):
    alpha_list = []
    for prime, exponent in prime_factors.items():
        if (prime - 1) % 4 == 0:
            alpha_list.append(exponent)
    return alpha_list

n = int(input())
alpha_list = extract_alpha_list(prime_factorization(n))
total_combinations = 1
for exponent in alpha_list:
    total_combinations *= (exponent * 2 + 1)
result = int(0.5 * (total_combinations - 1))
print(result)
```
